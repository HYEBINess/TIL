# Chunksize로 불러오기

```python
%%time
# chunksize로 가져오기
df = pd.read_csv('/Users/biny/Desktop/workspace/practice/data/초격차/crime.csv',chunksize=1000)
```

```python
df
```

```python
# chunksize로 불러온 애들은 dataframe이 아니기 때문에 info를 찍어볼 수 없다.  
df.info()
```

```python
%%time
sum_ward = 0
for i, chunk in enumerate(df):
  if i % 10000 == 0: # chunkrk 10000개씩 될때 마다 보기
    print(f"{i+1}th execution")
  sum_ward += chunk['Ward'].sum()
print(sum_ward)
```

# Dask Dataframe
* dask는 패키지 이름이다.

```python
from dask import dataframe as ddf
```

```python
%%time
ddf = ddf.read_csv('/Users/biny/Desktop/workspace/practice/data/초격차/crime.csv', dtype={'Community Area': 'float64', 
       'IUCR': 'object',
       'Ward': 'float64'})
```

```python
# npartitions=28 의미 -> 데이터가 많아서 파티션을 28개로 나눠놨어. -> 이 데이터에 대한 메타 데이터를 생성해놓았다고 표현함
ddf
```

```python
ddf.info()
```

```python
# head를 찍으면 그때서야 데이터를 불러옴
ddf.head(10)
```

```python
%%time

# pandas보다 반은 빠름
ddf_sum = ddf['Ward'].sum().compute()
```

```python
ddf_sum
```

```python
# 뒤에 compute()를 붙여주지 않으면 값이 안나옴. sum을 할 준비만 함 
ddf['Ward'].sum()
```
